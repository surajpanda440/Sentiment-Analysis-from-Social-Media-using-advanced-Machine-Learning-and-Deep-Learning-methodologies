```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.metrics import confusion_matrix, classification_report
import nltk

# Download the VADER lexicon
nltk.download('vader_lexicon')
```

```python
# Load the dataset
df = pd.read_csv('C:/Users/user/OneDrive/Desktop/Reviews.csv')  # Adjust your path

# Check the shape of the dataset
print("Original Data Shape:", df.shape)

# Normalize the text
df['Text'] = df['Text'].astype(str).str.lower()

# Prepare labels based on the Score column (1 for positive, 0 for negative)
df['Sentiment'] = df['Score'].apply(lambda x: 1 if x > 3 else 0)

# Optional: Check the first few rows of the dataset
print(df.head())
```

```python
print("Original Data Shape:", df.shape)
```

```python
# Initialize VADER Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Analyze Sentiments
def analyze_sentiment(text):
    score = analyzer.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 1  # Positive
    elif score['compound'] <= -0.05:
        return 0  # Negative
    else:
        return 0.5  # Neutral (optional)

# Apply sentiment analysis to the entire dataset
df['Predicted_Sentiment'] = df['Text'].apply(analyze_sentiment)
```

```python
# Convert predicted sentiments to binary
df['Predicted_Sentiment'] = df['Predicted_Sentiment'].apply(lambda x: 1 if x == 1 else 0)

# True labels
y_true = df['Sentiment']
y_pred = df['Predicted_Sentiment']

# Calculate evaluation metrics
print(classification_report(y_true, y_pred))

# Calculate accuracy
accuracy = np.mean(y_pred == y_true) * 100
print(f"Accuracy: {accuracy:.2f}%")
```

```python
# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

# Plotting the confusion matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Sample sentence to analyze
sample_text = "the worst day of my life"

# Get the VADER sentiment scores
sample_score = analyzer.polarity_scores(sample_text)

# Display the sentiment scores
print(f"Sentiment Scores for '{sample_text}':", sample_score)

# Generate heatmap for sentiment scores
sns.heatmap(pd.DataFrame(sample_score, index=[0]), annot=True, cmap='coolwarm', center=0)
plt.title(f'Sentiment Scores for: "{sample_text}"')
plt.show()
```
